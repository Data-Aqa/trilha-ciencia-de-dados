```python
# Importando bibliotecas
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
import numpy as np

# Criando objetos .Figure e .Axes
fig, ax = plt.subplots(figsize = (10,5))
ax.set_aspect('equal', adjustable = 'box')

# Criando pontos para o grfico
x = np.linspace(0, 4*np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Definindo tamanho dos eixos (x/y axis)
ax.set_ylim([-3, 3])
ax.set_xlim([0, 4*np.pi])

# Posio e label dos ticks
# X-Axis
major_x_ticks = np.arange(0, 4.5*np.pi, .5*np.pi)
minor_x_ticks = np.arange(0, 4.25*np.pi, .25*np.pi)
major_x_tick_labels = [r'0', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$', r'$\frac{5\pi}{2}$', r'$3\pi$', r'$\frac{7\pi}{2}$', r'$4\pi$']


# Plotando as duas linhas
ax.plot(x, y_sin, label = 'Seno')
ax.plot(x, y_cos, label = 'Cosseno')


# Definindo ttulos
ax.set_title('Titulo do Eixo (axes)', fontsize = 15, loc = 'center')
ax.set_xlabel('Titulo do eixo (axis) x')
ax.set_ylabel('Ttulo do eixo (axis) y')


# Ticks
# X-Axis
ax.set_xticks(major_x_ticks)
ax.set_xticks(minor_x_ticks, minor = True)
ax.set_xticklabels(major_x_tick_labels, fontdict = {'fontsize':11.75})
ax.tick_params(axis = 'x', length = 8.5)
ax.tick_params(axis = 'x', which = 'minor', size = 5)

# Y-Axis
ax.tick_params(axis = 'y', length = 8.5)

# Grids
ax.grid(alpha = .25, ls = '-.')

# Legendas e margens
ax.legend()
ax.margins(0)


# ANOTAES (Annotations)

# Funo para incluir crculos
def circler(text, xy, xytext):
    
    circulo = Circle(xy = xy, radius = .2, linewidth = 1, clip_on = False, facecolor = 'none', edgecolor = 'black')
    ax.add_artist(circulo)
    ax.annotate(text, xy = xytext, annotation_clip = False)
    
# Padro das setas
setas = dict(facecolor = '#828385', edgecolor = 'black', width = 1, headwidth = 8)

# Legenda
ax.annotate('legenda',
            xy = (3.35*np.pi, 2.1), 
            xytext = (2.75*np.pi, 2),
            arrowprops = setas)

# Plot 1
circler(text = 'plot 1', xy = (.5*np.pi, np.sin(.5*np.pi)), xytext = (.55*np.pi, 1.3))

# Plot 2
circler('plot 2', (2.5*np.pi, np.cos(2.5*np.pi)), (2.55*np.pi, 0.25))

# Grid
circler('grid', xy = (1.5*np.pi, 2), xytext = (1.55*np.pi, 2.3))

# Major xtick label
circler('major tick label', (np.pi, -3.45), (2.3, -4))

# Minor xtick
circler('minor xtick', xy = (2.25*np.pi, -3.05), xytext = (2.3*np.pi, -2.7))

# Major ytick
circler('major ytick', xy = (-0.05, -2), xytext = (0.3, -1.9))

# Spines
circler('spines', xy = (0, 3), xytext = (0.2, 3.15))

# Figura
circler('figura', xy = (3*np.pi, 3.8), xytext = (3.05*np.pi, 4.07))


# Alteraes a nvel da figura
# Ttulo da Figura
fig.suptitle('Ttulo da Figura', fontsize = 16)

# Ajuste da posio do subplot na figura
fig.subplots_adjust(top = .8)

# Cor da figura
fig.set_facecolor('#CFE0FF')


plt.show()
```

```python
fig.savefig('elementos.png', facecolor=fig.get_facecolor(), edgecolor='none', bbox_inches = 'tight')
```
